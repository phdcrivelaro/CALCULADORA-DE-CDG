```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Definindo os símbolos das moedas
usd_brl = 'USDBRL=X'
eur_brl = 'EURBRL=X'

# Calculando a data de início para os últimos 6 meses
end_date = pd.Timestamp.now()
start_date = end_date - pd.DateOffset(months=6)

# Importando os dados
data_usd = yf.download(usd_brl, start=start_date, end=end_date)
data_eur = yf.download(eur_brl, start=start_date, end=end_date)

# Criando um DataFrame para combinar os dados de fechamento ajustado
data_combined = pd.DataFrame({
    'USD': data_usd['Adj Close'],
    'EUR': data_eur['Adj Close']
})

# Calculando a correlação entre USD e EUR
correlation = data_combined.corr().iloc[0, 1]

# Medidas de tendência central e outras estatísticas
description = data_combined.describe()
range_stats = data_combined.max() - data_combined.min()
kurtosis = data_combined.kurtosis()
skewness = data_combined.skew()

# Plotando histogramas
data_combined.hist(bins=30, alpha=0.5)
plt.show()

# Imprimindo resultados
print("Coeficiente de correlação entre USD e EUR:", correlation)
print(description)
print("Amplitude:", range_stats)
print("Curtose:", kurtosis)
print("Skewness:", skewness)
```
